```python
trimmed_sales_data = sales_data[['Region', 'Country', 'Item Type', 'Total Revenue', 'Total Cost']]
```

```python
trimmed_sales_data
```

```python
regionwise_sales = trimmed_sales_data.groupby('Region').sum()
```

```python
regionwise_sales
```

```python
regionwise_sales.plot.bar()
```

```python
regionwise_sales.plot.bar(stacked=True)
```

```python
itemwise_sales = trimmed_sales_data.groupby('Item Type').sum()
```

```python
itemwise_sales.plot.bar()
```

```python
countrywise_sales = trimmed_sales_data.groupby('Country').sum()
```

```python
countrywise_sales.plot.bar()
```

```python
top_cwise_sales = countrywise_sales.sort_values(by='Total Revenue', ascending=False)
```

```python
top_cwise_sales
```

```python
top_ten_cdata = top_cwise_sales[0:10]
```

```python
top_ten_cdata.plot.bar(stacked=True)
```
